```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

Everst_data=pd.read_csv('average_premium.csv')
data=Everst_data
data['Attained_age']=data['Age']+data['Duration']
data['Premium_rate']=data['BasePremiumAmount']/data['BaseNARAmount']*10000
#exclude_cols = ['Gender','RiskClass']
Coulmns_to_keep=['Age','Premium_rate','Gender','RiskClass','Duration']
#data=data[data['SingleOrJointType']=="Single"]
#data=data[data['Gender']=="Male"]
#data=data[data['RiskClass']=="NS1"]
data=data[Coulmns_to_keep]
data=data[(data['RiskClass']!="SS1") & (data['RiskClass']!="SS2")& (data['RiskClass']!="SP1")]
data['index']=data.index
numerical_columns = data.select_dtypes(include=['number']).columns
categorical_columns = data.select_dtypes(include=['object']).columns
data_copy=data.copy()
data
```

```python
# Create a multi-index dataframe using Age, Duration, and a combined Gender and RiskClass column

import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
data['RiskClass'] = data['RiskClass'].replace({'N+1': 'NP2', 'NP1': 'NP2'}).replace({'NS3': 'NS2', 'NS2': 'NS2', 'NR1': 'NS2', 'NS1' : 'NS2'})
print(data)
print(data['RiskClass'].unique())
multi_index_df = data.set_index(['Age', 'Duration', data['Gender'] + '_' + data['RiskClass']])

# Convert the multi-index dataframe to a regular dataframe
result_df = multi_index_df.reset_index().drop(['Gender', 'RiskClass'], axis=1)
print(result_df)
result_df['level_2'].unique()
```

```python
import numpy as np
import pandas as pd

# Before removing NAs
print("Before removing NAs:")
print(result_df.describe(include='all'))
print(result_df.count())

# Drop rows with missing values
result_df_no_na = result_df.dropna()

# After removing NAs
print("\nAfter removing NAs:")
print(result_df_no_na.describe(include='all'))
print(result_df_no_na.count())
```

```python
# prompt: for teh data ccreate the difference of value of premium rate with the avearege in each frame where frameis oof length  age 5 and duration lengthe of 3

import numpy as np
import pandas as pd

# Assuming 'result_df_no_na' is your DataFrame from the preceding code

# Define frame sizes
age_frame_size = 5
duration_frame_size = 3

# Create a new column for the difference from average premium rate
result_df_no_na['Premium_Diff'] = 0

# Iterate over age and duration frames
for age_start in range(result_df_no_na['Age'].min(), result_df_no_na['Age'].max() + 1, age_frame_size):
  for duration_start in range(result_df_no_na['Duration'].min(), result_df_no_na['Duration'].max() + 1, duration_frame_size):
    age_end = age_start + age_frame_size
    duration_end = duration_start + duration_frame_size

    # Filter data for the current frame
    frame_data = result_df_no_na[
        (result_df_no_na['Age'] >= age_start) & (result_df_no_na['Age'] < age_end) &
        (result_df_no_na['Duration'] >= duration_start) & (result_df_no_na['Duration'] < duration_end)
    ]

    # Calculate the average premium rate for the frame
    avg_premium = round(frame_data['Premium_rate'].mean(),2)

    # Calculate the difference from average for each row in the frame
    result_df_no_na.loc[frame_data.index, 'Premium_Diff'] = round(frame_data['Premium_rate'] - avg_premium,2)

print(result_df_no_na.sort_values(by =['level_2','Age','Duration']))
```

```python
# Calculate average Premium_Diff for duplicates
average_premium_diff = result_df_no_na.groupby(['Age', 'Duration', 'level_2'])['Premium_Diff'].mean().reset_index()

average_premium_diff.pivot(index=['level_2','Age'], columns='Duration', values='Premium_Diff')
average_premium_diff.to_csv('average_premium_diff.csv')
```

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import KNNImputer

# One-hot encode 'level_2'
average_premium_diff_encoded =pd.get_dummies(average_premium_diff, columns=['level_2'], prefix=['level_2']) #average_premium_diff #


# Create a complete grid of Age and Duration values
all_ages = range(average_premium_diff_encoded['Age'].min(), average_premium_diff_encoded['Age'].max())
all_durations = range(average_premium_diff_encoded['Duration'].min(), 10)
all_combinations = pd.MultiIndex.from_product([all_ages, all_durations], names=['Age', 'Duration'])

# Reindex the DataFrame to include all combinations and fill missing values with NaN
# Drop duplicates before reindexing to ensure a unique MultiIndex
average_premium_diff_encoded = average_premium_diff_encoded.drop_duplicates(subset=['Age', 'Duration'])
average_premium_diff_filled = average_premium_diff_encoded.set_index(['Age', 'Duration']).reindex(all_combinations).reset_index()

# Impute missing values using KNNImputer
imputer = KNNImputer(n_neighbors=5)
average_premium_diff_imputed = pd.DataFrame(imputer.fit_transform(average_premium_diff_filled), columns=average_premium_diff_filled.columns)
#average_premium_diff_encoded.pivot(index=['level_2','Age'], columns='Duration', values='Premium_Diff')
```

```python
# prompt: this is a  age*duration*level_2_Female_NS1  dataframe of (75*10*2) , use a cnn of filter and frame of 5*3*2 to predict the Average absolute value in the frame to detect anomalies in the premium diff

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv3D, Flatten, Dense, Reshape

# Reshape the data for CNN input
# Assuming 'average_premium_diff_encoded' is your DataFrame
average_premium_diff_array = average_premium_diff_encoded.drop(['Age', 'Duration'], axis=1).values.reshape(75, 10, 2, 1)
```

```python

# Create the CNN model
model = Sequential()
model.add(Conv3D(1, (5, 3, 2), activation='relu', input_shape=(75, 10, 2, 1)))
model.add(Flatten())
model.add(Dense(1))  # Output layer predicting average absolute value

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model (you'll need to split your data into training and validation sets)
# model.fit(X_train, y_train, epochs=..., validation_data=(X_val, y_val))

# Predict on the entire dataset
predictions = model.predict(average_premium_diff_array)

# Reshape predictions to match original DataFrame
predictions = predictions.reshape(75, 10)

# Calculate absolute values of predictions
abs_predictions = np.abs(predictions)

# Add predictions to DataFrame
average_premium_diff_encoded['Predicted_Abs_Diff'] = abs_predictions.flatten()

# Identify anomalies (e.g., values above a certain threshold)
threshold = 0.1  # Adjust this threshold based on your data and expectations
anomalies = average_premium_diff_encoded[average_premium_diff_encoded['Predicted_Abs_Diff'] > threshold]

print(anomalies)
```
